```python
def correlation_heatmap(correlation):
    fig = plt.figure(figsize = (25,25))
    ax = fig.add_subplot(111)
    ax = sns.heatmap(
        correlation, 
        vmin=-1, vmax=1, center=0,
        cmap=sns.diverging_palette(20, 220, n=200),
        square=True
    )
    ax.set_xticklabels(
        ax.get_xticklabels(),
        rotation=45,
        horizontalalignment='right'
    )
    plt.show()
    
correlation = df.corr()
correlation_heatmap(correlation)
```

```python
df.dropna(axis =1, inplace=True)
df.head()
```

```python
correlation = df.corr()
correlation_heatmap(correlation)
```

```python
def standardize(df):
    columns = df.columns
    for i in columns:
        mean = np.mean(df[i])
        std = np.std(df[i])
        df[i] = (df[i] - mean) / std
    return df
standardize(df)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=15)
pca.fit(df)
```

```python
pca.explained_variance_ratio_.sum()
```

```python
df_new = pca.fit_transform(df)
df_new = pd.DataFrame(df_new,columns=[i+1 for i in range(15)])
df_new.head()
```

```python
columns = df_new.columns
columns_num = len(columns) 

for i in range(columns_num):
    plt.figure(figsize=(10,2))
    plt.plot(df_new[columns[i]],label = columns[i],c=np.random.rand(3,))
    plt.legend()
    plt.show()
```

```python
df_new['Velocity'] = df['Velocity']
correlation_heatmap(df_new.corr())

df_new.to_csv('new_data.csv',index = False)
```

```python
import torch
from torch import nn
from torch.autograd import Variable
 
class RNN(nn.Module):
    def __init__(self):
        super(RNN, self).__init__()
 
        self.rnn = nn.RNN(
            input_size=15,
            hidden_size=32,
            num_layers=2,
            batch_first=True,
        )
        self.linear = nn.Linear(32, 1)
 
    def forward(self, x, h_n): 
        r_out, h_n = self.rnn(x, h_n)
 
        outs = []
        for step in range(r_out.size(1)):
            outs.append(self.linear(r_out[:, step, :]))

net = RNN()

```
